```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set(color_codes=True)
df = pd.read_excel("ipqdata.xls")
data = df[(np.abs(stats.zscore(df)) < 3).all(axis=1)]
data_long = pd.melt(data, "INTERFACE", var_name="IPQ", value_name="value")
sns.factorplot("IPQ", hue="INTERFACE", y="value", data=data_long, kind="box", notch=True)
plt.show()
```
